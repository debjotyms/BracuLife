```python
#Task_1
def bbl(my_list):
  for i in range(len(my_list)-1):
    for j in range(len(my_list)-1-i):
      if(my_list[j]>my_list[j+1]):
        temp = my_list[j]
        my_list[j] = my_list[j+1]
        my_list[j+1] = temp
  return my_list
my_list = [10, 1, 20, 3, 6, 2, 5, 11, 15, 2, 12, 14, 17, 18, 29]
my_list = bbl(my_list)
print(my_list)
```

```python
#Task_8
lst = input()
lst = lst[1:-1]
lst = [int(i) for i in lst.split(',')]
lind = 0
rind = 0
min = float('inf')
for i in range(len(lst)-1):
  for j in range(i+1,len(lst)):
    a = abs(lst[i]+lst[j])
    if(a<min):
      min = a
      lind = i
      rind = j
print(f'Two pairs which have the smallest sum = {lst[lind]} and {lst[rind]}')
```

```python
#Task_9
import math
points = [(5,3), (2,9), (-2,7), (-3,-4), (0,6), (7,-2)]
lsti = 0
min = float('inf')
for i in range(len(points)):
  dist = math.sqrt(math.pow(points[i][0],2)+math.pow(points[i][1],2))
  if(dist<min):
    min = dist
    lsti = i
print(f'Minimum distance = {min}')
print(f'Here the closest point is {points[lsti]} which has a distance of {min} from the origin.')
```
