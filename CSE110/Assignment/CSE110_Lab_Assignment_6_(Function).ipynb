```python
#task_2
def fibonacci(e):
  a = 0
  b = 1
  print(a,end=' ')
  while b<=e:
    print(b,end=' ')
    c = a
    d = b
    a = d
    b = c+d
    if(b>e):
      break
fibonacci(int(input()))
```

```python
#Task_7
def show_palindrome(a):
  s = ""
  for i in range(1,a+1):
    s = s+str(i)
  for i in range(a-1,0,-1):
    s = s+str(i)
  return s
print(show_palindrome(5))
```

```python
#Task_8
def show_palindromic_triangle(a):
  for i in range(1,a+1):
    palind2 = show_palindrome(i)
    palind = ''
    for j in palind2:
      palind += j + ' '
    palind = ((2*(5-i))*' ')+palind
    print(palind)
show_palindromic_triangle(5)
```

```python
#Task_9
import math

def area_circumference_generator(a):
  tpl = []
  tpl.append(math.pi*a*a)
  tpl.append(2*math.pi*a)
  return tpl

tpl = area_circumference_generator(int(input()))
a,b = tpl
print(f"Area of the circle is {a} and circumference is {b}")
```

```python
#Task_10
def make_square(a):
  dc = {}
  for i in range(a[0],a[-1]+1):
    dc[i] = i*i
  return dc
print(make_square((5,9)))
```

```python
#Task_11
def rem_duplicate(a):
  lst = []
  for i in a:
    if(i not in lst):
        lst.append(i)
  return tuple(lst)
print(rem_duplicate((1,1,1,2,3,4,5,6,6,6,6,4,0,0,0)))
```

```python
#Task_12
def function_name(a):
  lst = []
  for i in a:
    if(lst.count(i)<2):
        lst.append(i)
  b = len(a)-len(lst)
  print(f"Removed:{b}")
  return (lst)
print(function_name([1,2,3,3,3,3,4,5,8,8]))
```

```python
#Task_13
def calc_func(a,b,c):
  if(a=='+'):
    return b+c
  elif(a=='-'):
    return b-c
  elif(a=='*'):
    return b*c
  else:
    return b/c
print(calc_func("*",2,3))
```
